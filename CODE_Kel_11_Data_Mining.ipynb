```python
import numpy as np

def map_hour_to_illumination(hour):
    if 5 <= hour <= 6:
        return 'DAWN'
    elif 7 <= hour <= 17:
        return 'DAYLIGHT'
    elif 18 <= hour <= 18:
        return 'DUSK'
    else:
        return 'DARK - LIGHTED'

df_selected['Illumination Description'] = df_selected.apply(
    lambda row: map_hour_to_illumination(row['Hour']) if pd.isna(row['Illumination Description']) else row['Illumination Description'],
    axis=1
)

print(df_selected['Illumination Description'].value_counts(dropna=False))
```

Missing value pada kolom '*Illumination Description*' berdasarkan jam dari '*Hour*'.
Maka secara logika:

05:00 – 06:59 → DAWN (subuh)

07:00 – 17:59 → DAYLIGHT (siang)

18:00 – 18:59 → DUSK (senja)

19:00 – 04:59 → DARK (malam)

```python
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

numerik_fitur = ['Lat', 'Long', 'x', 'y', 'Number of Motor Vehicles', 'Number of Injuries', 'Number of Fatalities']
print(df_encoded[numerik_fitur].describe())

for col in numerik_fitur:
    plt.figure(figsize=(6,4))
    df_encoded[col].hist(bins=30)
    plt.title(f'Distribusi {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

    skew_value = skew(df_encoded[col].dropna())
    kurt_value = kurtosis(df_encoded[col].dropna())

    print(f"{col}: Skewness = {skew_value:.2f}, Kurtosis = {kurt_value:.2f}")

    if abs(skew_value) < 0.5:
        print(f"{col} : Distribusi Normal (Skewness dekat 0)")
    else:
        print(f"{col} : Distribusi Miring (Skewness {skew_value:.2f})")
```

```python
from sklearn.impute import SimpleImputer

imputer_mean = SimpleImputer(strategy='mean')
imputer_median = SimpleImputer(strategy='median')

mean_features = ['Number of Motor Vehicles']
median_features = ['Lat', 'Long', 'x', 'y', 'Number of Injuries', 'Number of Fatalities']

df_encoded[mean_features] = imputer_mean.fit_transform(df_encoded[mean_features])
df_encoded[median_features] = imputer_median.fit_transform(df_encoded[median_features])

print(df_encoded[mean_features + median_features].isnull().sum())
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Data Preparation (seperti sebelumnya) ---
illumination_summary = df_selected.groupby('Illumination Description').agg(
    accident_count=('Illumination Description', 'size'),
    total_injuries=('Number of Injuries', 'sum'),
    total_fatalities=('Number of Fatalities', 'sum')
).reset_index()

illumination_summary = illumination_summary[
    ~illumination_summary['Illumination Description'].str.contains('unknown|other', case=False)
]

illumination_summary['total_victims'] = (
    illumination_summary['total_injuries'] + illumination_summary['total_fatalities']
)

illumination_labels = illumination_summary['Illumination Description']
colors = plt.cm.Paired(range(len(illumination_summary)))

# Fungsi untuk menambahkan label di luar segmen
def add_labels(ax, wedges, values):
    total = sum(values)
    for i, wedge in enumerate(wedges):
        angle = (wedge.theta2 + wedge.theta1) / 2
        x = np.cos(np.deg2rad(angle))
        y = np.sin(np.deg2rad(angle))
        percentage = f"{(values[i]/total)*100:.1f}%"
        ax.annotate(percentage, xy=(x, y), xytext=(1.1*x, 1.1*y),
                    ha='center', va='center', fontsize=10,
                    arrowprops=dict(arrowstyle="-", color='gray'))

# --- Plotting ---
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# Donut Chart: Jumlah Kecelakaan
wedges1, _ = axes[0].pie(
    illumination_summary['accident_count'],
    colors=colors,
    startangle=90,
    wedgeprops=dict(width=0.3, edgecolor='white')
)
centre_circle1 = plt.Circle((0, 0), 0.70, fc='white')
axes[0].add_artist(centre_circle1)
axes[0].set_title('Resiko kecelakaan dengan Cedera\nBerdasarkan Kondisi Penerangan', fontsize=14)
axes[0].axis('equal')
add_labels(axes[0], wedges1, illumination_summary['accident_count'])

# Donut Chart: Jumlah Korban
wedges2, _ = axes[1].pie(
    illumination_summary['total_fatalities'],  # Fokus ke jumlah kematian
    colors=colors,
    startangle=90,
    wedgeprops=dict(width=0.3, edgecolor='white')
)
centre_circle2 = plt.Circle((0, 0), 0.70, fc='white')
axes[1].add_artist(centre_circle2)
axes[1].set_title('Resiko Kematian\nBerdasarkan Kondisi Penerangan', fontsize=14)
axes[1].axis('equal')
add_labels(axes[1], wedges2, illumination_summary['total_fatalities'])

# Legend satu kali saja
fig.legend(wedges1, illumination_labels,
           title="Kondisi Pencahayaan",
           loc="center right",
           bbox_to_anchor=(1.15, 0.5))

plt.tight_layout()
plt.show()
```

```python
df_scaled['tingkat_keparahan'] = (df_scaled['Number of Injuries'] > 1).astype(int)
print(df_scaled['tingkat_keparahan'].value_counts())
```

Klasifikasi

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cyclical_features = ['Month', 'Hour', 'Day']
X = df_scaled.drop(columns=['tingkat_keparahan', 'Number of Injuries'] + cyclical_features)
y = df_scaled['tingkat_keparahan']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, random_state=42, test_size=0.2
)

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight='balanced',
    n_jobs=-1
)

rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
labels = rf.classes_

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix - Random Forest")
plt.tight_layout()
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=4))
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cyclical_features = ['Month', 'Hour', 'Day']
X = df_scaled.drop(columns=['tingkat_keparahan', 'Number of Injuries'] + cyclical_features)
y = df_scaled['tingkat_keparahan']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train_res, y_train_res)
y_pred = rf.predict(X_test)
cm = confusion_matrix(y_test, y_pred, labels=rf.classes_)

plt.figure(figsize=(6, 5))
sns.heatmap(
    cm,
    annot=True, fmt='d', cmap='Blues',
    xticklabels=rf.classes_, yticklabels=rf.classes_
)
plt.title("Confusion Matrix – RF + SMOTE", fontsize=14)
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.tight_layout()
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=4))
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

cyclical_features = ['Month', 'Hour', 'Day']
X = df_scaled.drop(columns=['tingkat_keparahan', 'Number of Injuries'] + cyclical_features)
y = df_scaled['tingkat_keparahan']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

xgb = XGBClassifier(
    random_state=42,
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    eval_metric='mlogloss',
    n_jobs=-1
)

xgb.fit(X_train, y_train)

y_pred = xgb.predict(X_test)

cm = confusion_matrix(y_test, y_pred, labels=xgb.classes_)

plt.figure(figsize=(6, 5))
sns.heatmap(
    cm,
    annot=True, fmt='d', cmap='YlOrRd',
    xticklabels=xgb.classes_,
    yticklabels=xgb.classes_
)
plt.title("Confusion Matrix – XGBoost Classifier", fontsize=14)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.tight_layout()
plt.show()

print("\nClassification Report:")
print(classification_report(y_test, y_pred, digits=4))
```
